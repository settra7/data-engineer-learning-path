# Kaggle Dataset Analytics Pt.2

```python
import pandas as pd
import numpy as np
import os
file_path = os.path.abspath('') + "\\" + "W02_D7_titanic.csv"
df = pd.read_csv(file_path)
```

```python
# Display top 10 most aged passengers
df.sort_values("Age", ascending=False).head(10)
```

```python
# Rank passengers by age (oldest to youngest) and display top 10
df['Rank_Age'] = df['Age'].rank(ascending=False, method='dense')
df.sort_values("Rank_Age").head(10)
```

```python
# Calculate survival rates for adults
df[df['Age'] >= 18].groupby('Sex')['Survived'].mean()
```

```python
# Create a pivot table to analyze survival
pd.pivot_table(df, values="Survived", index="Sex", columns="Pclass", aggfunc="mean")
```

```python
# Save the pivot table to a CSV file
df2 = pd.pivot_table(df, values="Survived", index="Sex", columns="Pclass", aggfunc="mean")
df2.to_csv("W02_D7_titanic_pivot.csv")
```
